```python
#both measurements are normal distributed
lisa = 110
susan = 90
sd =5
firstmeasurement = 115
#find the expected value of second mearsurement 
```

```python
import random
import numpy as np
```

```python
n=10000
mylist=[]
for i in range(n):
    if random.randint(0,1)<0.5:
        x=np.random.normal(lisa, sd, 1)
    else:
        x=np.random.normal(susan, sd, 1)
    mylist.append(x.item())
```

```python
import pandas as pd
mylist=pd.Series(mylist)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig,ax=plt.subplots(figsize=(10,10))
mylist.plot(kind="kde")
```
